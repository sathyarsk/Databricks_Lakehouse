```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

# Business Transformation and Modelling

```python
query="""
select row_number() over(order by customer_id) as customer_key,
customer_id,
customer_name,
segment,
country,
city
from silver.customer
"""
df=spark.sql(query)
```

# Write it to Gold Tabe

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("gold.dim_customer")
)
```

```python
%sql
select * from gold.dim_customer;
```
